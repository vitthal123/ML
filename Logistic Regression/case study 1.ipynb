```python
# a. Make a bar plot with each state name on the x -axis and their total 
# benefitted inmates as their bar heights. Which state has the maximum 
# number of beneficiaries?
# [Hint: Use bar method of pyplot]

import matplotlib.pyplot as plt
plt.figure(figsize=(30,10))
plt.bar(x='STATE/UT',height='total_benefitted',data=df_prisoners)
```

```python
labels = df_prisoners["STATE/UT"]
values = df_prisoners["total_benefitted"]
plt.figure(figsize=(80,20))
plt.pie(values, labels = labels, autopct='%1.1f%%', shadow=True)
plt.title("STATE/UT vs Total_inmates_benefitted")
plt.show()
```
