```python
## Loading Libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
## Loading Data
df_movie=pd.read_csv('movies.csv')
df_ratings=pd.read_csv('ratings.csv')

df_movie.head()
```

```python
## Merfing movie with ratings
df_user_ratings=pd.merge(df_ratings,df_movie,on='movieId')
df_user_ratings
```

```python
user_rating_mat=df_user_ratings.pivot_table(index='userId',columns='title',values='rating',aggfunc='sum')
 
```

```python
user_rating_mat
```

```python
user_rating_mat.shape
```

```python
user_ratings=user_rating_mat.dropna(thresh=10,axis=1).fillna(0,axis=1)
```

```python
user_ratings.shape
```

```python
user_ratings.head()
```

```python
corr_matrix=user_ratings.corr(method='pearson')
```

```python
corr_matrix.head()
```

```python
def get_similar_movies(movie_name,ratings):
    similar_ratings=corr_matrix[movie_name]
    similar_ratings=similar_ratings.sort_values(ascending=False)[:5]
    return similar_ratings
```
