```python
## Dataset Creation

from  sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import matplotlib.style as style

```

```python
plt.grid()
X,y=make_blobs(n_samples=100,n_features=2,centers=3)
plt.scatter(X[:,0],X[:,1],s=40,color='g')
plt.xlabel('X- Axis')
plt.ylabel('Y- Axis')
plt.show()
```

```python
## Creating Datasets with moon
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
from matplotlib import style 
```

```python
X,y=make_moons(n_samples=100,noise=0.01)
plt.scatter(X[:,0],X[:,1],s=40,color='r')
plt.xlabel('X- Axis')
plt.ylabel('Y- Axis')
plt.show()
```

```python
## Creating Normal Ditribution
## define mu(mean)
## standard deviation : sigma
mu=0.5
sigma=0.1
## starting value is 0
np.random.seed(0)
X=np.random.normal(mu,sigma,(395,1))
```

```python
# define the y co-ordinates
Y = np.random.normal(mu * 2, sigma * 3, (395, 1))
 
# plot a graph
plt.scatter(X, Y, color = 'g')
plt.show()
```
