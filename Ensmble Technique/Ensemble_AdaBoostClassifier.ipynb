```python
l=['column0','column1','column2','column3','column4','column5','column6','column7','column8','column9','column10','column11','column12',
 'column13','column14','column15','column16','column17','column18','column19','column20','column21','column22','column23','column24',
 'column25','column26','column27','column28','column29','column30','column31','column32','column33','column34','column35','column36',
 'column37','column38','column39','column40','Label']
```

```python
df_data=pd.read_csv('bio-degradabale-data.csv',sep=';',names=l)
df_data.head()
```

```python
df_data['Label'].unique()
```

```python
# label_enc=LabelEncoder()
# df_data['Label']=label_enc.fit_transform(df_data['Label'])
df_data['Label']=df_data['Label'].map({'RB':0,'NRB':1})
```

```python
df_data.head()
```

```python
x=df_data.iloc[:,:-1]
y=df_data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy',max_depth=1)
# adaboost=
```

```python
adaboost=AdaBoostClassifier(base_estimator=model,n_estimators=400,learning_rate=1)
best_model=adaboost.fit(x_train,y_train)
```

```python
y_pred=best_model.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```
